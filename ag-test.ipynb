```python
import random

TARGET_STRING = "Hello, Genetic Algorithm!"
POPULATION_SIZE = 100
MUTATION_RATE = 0.01
```

```python
def generate_individual(length):
    return ''.join(random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz ,.!?") for _ in range(length))

def generate_population(population_size, target_string):
    return [generate_individual(len(target_string)) for _ in range(population_size)]
```

```python
def fitness(individual, target_string):
    return sum(c1 == c2 for c1, c2 in zip(individual, target_string))
```

```python
def selection(population, target_string):
    return max(population, key=lambda individual: fitness(individual, target_string))
```

```python
def crossover(parent1, parent2):
    split_point = random.randint(0, len(parent1) - 1)
    child = parent1[:split_point] + parent2[split_point:]
    return child
```

```python
def mutate(individual, mutation_rate):
    mutated_individual = list(individual)
    for i in range(len(mutated_individual)):
        if random.random() < mutation_rate:
            mutated_individual[i] = random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz ,.!?")
    return ''.join(mutated_individual)
```

```python
def genetic_algorithm(target_string, population_size, mutation_rate, generations):
    population = generate_population(population_size, target_string)

    for generation in range(generations):
        population = sorted(population, key=lambda individual: fitness(individual, target_string), reverse=True)
        fittest_individual = selection(population, target_string)

        if fitness(fittest_individual, target_string) == len(target_string):
            print(f"Generation {generation}: Found solution - {fittest_individual}")
            break

        print(f"Generation {generation}: Best fitness - {fitness(fittest_individual, target_string)}")
        print(f"Best individual - {fittest_individual}\n")

        new_population = [fittest_individual]

        while len(new_population) < population_size:
            parent1 = random.choice(population)
            parent2 = random.choice(population)
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            new_population.append(child)

        population = new_population

if __name__ == "__main__":
    genetic_algorithm(TARGET_STRING, POPULATION_SIZE, MUTATION_RATE, generations=10000)
```
